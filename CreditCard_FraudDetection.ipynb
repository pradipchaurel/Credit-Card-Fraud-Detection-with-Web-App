```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

```python
# read the dataset
data = pd.read_csv('creditcard.csv')
data.head()
```

```python
# describe method generated a statistical summary for numerical columns.
data.describe()
```

```python
# analyzing classs --> class == 1 (fraud transaction) & class == 0 (valid transaction)
# it calculates the ration of fraud cases to valid cases

fraud = data[data['Class']== 1]  # data -->df
valid = data[data['Class'] == 0]

outlierFraction = len(fraud)/len(valid)
print(outlierFraction)
print('Fraud Cases:',(len(data[data['Class']==1])))
print('Valid Cases:',(len(data[data['Class']==0])))
```

```python
# Exploring fraud Transaction amounts
print('Amount details of the fraud transaction')
fraud.Amount.describe()
```

```python
# Exploring valid Transaction amounts
print('Amount details of the valid transaction')
valid.Amount.describe()
```

From the above we can see the average amounts for fraud transaction tend to have higer than valid transaction.

```python
# plotting correlation Matrix
'''Correlation matrix is a table that shows hoe different variables are related
to each other'''

corrmat = data.corr()
fig = plt.figure(figsize = (12,9))
sns.heatmap(corrmat,vmax = 0.8, square = True)
plt.show()
```

Data Preparation

```python
# separate the input features(X) and target variable(Y)
X = data.drop(['Class'],axis =1)
Y = data['Class']

print(X.shape)  # prints numbers of rows and columns
print(Y.shape)

xData = X.values    # it converts dataframe to numpy array for fast processing
yData = Y.values
```

Split the dataset into Training and Testing data

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest = train_test_split(xData, yData, test_size = 0.2,random_state = 42)
```

# Building and Training the model --

-Fraud detection datasets are usually highly imbalanced (e.g., 0.1%    fraud, 99.9% non-fraud).
-Random Forests can focus on the minority class using techniques like:
1.Class weights
2.Balanced subsampling

-It can help detect rare fraudulent cases without being overwhelmed by  the majority class.


```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(xTrain,yTrain)

yPred = rfc.predict(xTest)
```

# let's Evaluate the Model
After training the model we need to evaluate its performance using various metrics such as accuracy, precision, recall, F1-score and the Matthews correlation coefficient.

# Accuracy: Measures how often the model is correct overall.

# Precision: Measures how many predicted frauds are actually fraud.

# Recall: Measures how many actual frauds the model correctly found.

# F1-Score: Balances precision and recall into one score.

# Matthews Correlation Coefficient (MCC): Measures overall prediction  quality, even with imbalanced data.


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef 

accuracy = accuracy_score(yTest, yPred)
precision = precision_score(yTest, yPred)
recall = recall_score(yTest, yPred)
f1 = f1_score(yTest, yPred)
mcc = matthews_corrcoef(yTest, yPred)

print("--Model Evaluation Matrics--")
print(f"Accuracy:{accuracy:.4f}")
print(f"Precision:{precision:.4f}")
print(f"Recall:{recall:.4f}")
print(f"F1-Score:{f1:.4f}")
print(f"Matthews Correlation Coefficient:{mcc:.4f}")

```

```python
# Draw the confusion metrix
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(yTest,yPred)
plt.figure(figsize = (8,6))
sns.heatmap(conf_matrix,annot=True,fmt="d",cmap="Blues",xticklabels=['Normal','Fraud'],yticklabels=['Normal','Fraud'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```
